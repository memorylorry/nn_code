# 常见的图表

## 1. 散点图

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 制造特征值、标签
data = torch.ones(2000,2)
x1 = torch.normal(2*data,1)
x2 = torch.normal(6*data,1)
y1 = torch.zeros(len(x1))
y2 = torch.ones(len(x2))
# 制造标签
x = torch.cat((x1,x2),axis=0)
y = torch.cat((y1,y2),axis=0)
```

A [scatter](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.scatter.html#matplotlib.pyplot.scatter) plot of y vs. x with varying marker size and/or color.

| 参数名 | 说明 |
|----|----|
| s | 点的大小，可以输入`标量`或`矢量`(形状同输入数据)，默认为20 |
| c | 色彩颜色序列 |
| marker | 标记类型，默认为`o`，其它形状参考[matplotlib.markers](https://matplotlib.org/api/markers_api.html#module-matplotlib.markers) |

```python
plt.figure(figsize=(12,12))
plt.scatter(x[:,0],x[:,1],s=53,c=y, cmap=plt.cm.Paired,marker='o',alpha=0.2)
```

## 2. 直线

```python
x = np.arange(1,10,0.5)
y = np.exp(x)
plt.plot(x,y,'ro',label='e^x point')
plt.plot(x,y,'r-',label='e^x line')
plt.legend()
```

## 3. 柱状图

```python
# 单一的柱状图
x = np.random.rand(10)
labels = np.arange(0,10)
plt.bar(labels,height=x)
```

```python
# 多项目的柱状图
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 35, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
men_std = [2, 3, 4, 1, 2]
women_std = [3, 5, 2, 3, 3]
width = 0.35       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()

# 2个label的值一样，则是堆叠效果
ax.bar(labels, men_means, width, label='Men')
ax.bar(labels, women_means, width, label='Women')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.legend()
# fig.tight_layout()
```

```python
# 多项目的柱状图
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 35, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
x = np.arange(len(labels))  # the label locations
width = 0.35       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()

# 2个label的值一样，则是堆叠效果
ax.bar(x-width/2, men_means, width, label='Men')
ax.bar(x+width/2, women_means, width, label='Women')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.legend()
# fig.tight_layout()
```
